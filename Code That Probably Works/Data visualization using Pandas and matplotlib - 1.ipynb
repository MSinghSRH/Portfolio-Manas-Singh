```python
#importing arrays and plotting specific libraries, and linking the needed files
import matplotlib.pyplot as plt
import seaborn as sb
sampleset2 = pd.read_csv('/content/drive/MyDrive/Blender - Open Data.csv')
sampleset2
```

```python
#removing the "number of benchmarks column", as it is not needed, or relevant to the purpose of plotting
sampleset2edited = sampleset2.drop('Number of Benchmarks',axis=1)
sampleset2edited
```

```python
#trying out the first plot

#giving each column its own array
device_names = sampleset2edited['Device Name']
median_scores = sampleset2edited['Median Score']

#plotting the line plot
plt.plot(device_names, median_scores)
plt.show()
```

```python
#the older data set has device names that are too long, and were overlapping each other, making the text unintelligible. 
#The data set was modified in notepad to only show the relevant device names and then saved as a .csv file

#linking the new file
sampleset2 = pd.read_csv('/content/drive/MyDrive/Blender - Open Data2.csv')
sampleset2
```

```python
#plotting the chart again

#attaching the columns to their own arrays
device_names = sampleset2['Device Name']
median_scores = sampleset2['Median Score']

#plotting the chart
plt.plot(device_names, median_scores)
plt.show()
```

```python
#the overlapping problem persists. So we switched the axes.
device_names = sampleset2['Device Name']
median_scores = sampleset2['Median Score']

plt.plot(median_scores, device_names)
plt.show()
```

```python
#however, the resulting plot isn't very easy to read.
device_names = sampleset2['Device Name']
median_scores = sampleset2['Median Score']

#this time, while plotting, we rotated the device name labels by 90 degrees to get rid of the overlaps
plt.plot(device_names, median_scores)

#this is the code to rotate the labels, 'ha' here means "horizontal alignment"
plt.xticks(rotation=90, ha='center')
plt.show()
```

```python
#scatter plot was plotted next, and the labels were rotated, as in the previous chart
plt.scatter(device_names, median_scores)
plt.xticks(rotation=90, ha='center')
plt.show()
```

```python
#the data is more legible, but our data is sorted by ascending order of scores, and not generationally. 
#So, now, we sort the data by generation of devices

#Sorting sampleset2 by the 'Device Name' column
data_sorted = sampleset2.sort_values('Device Name')

#Extracting the sorted data to their own arrays
device_names = data_sorted['Device Name']
median_scores = data_sorted['Median Score']

#plotting
plt.plot(device_names, median_scores)
plt.xlabel('Device Name')
plt.ylabel('Median Score')
plt.title('Median Scores by Device (Sorted by Device Name)')
plt.xticks(rotation=90, ha='center')

plt.show()
```

```python
#plotting the new scatterplot

plt.scatter(device_names, median_scores)
plt.xlabel('Device Name')
plt.ylabel('Median Score')
plt.title('Median Scores by Device (Sorted by Device Name)')
plt.xticks(rotation=90, ha='center')
plt.show()
```

```python
#plotting the bar chart

plt.bar(device_names, median_scores)
plt.xlabel('Device Name')
plt.ylabel('Median Score')
plt.title('Median Scores by Device (Sorted by Device Name)')
plt.xticks(rotation=90, ha='center')
plt.show()
```

```python
#Although plotting a histogram will make no sense here, as a histogram needs cumulative frequencies to plot the distribution. 
#We plotted it anyway to see how matplotlib.pyplot interprets the data

plt.hist(median_scores)
plt.ylabel('Device Name')
plt.xlabel('Median Score')
plt.show()
```

```python
#similarly, the same was done with a pie chart to see how its interpreted by the library.

plt.pie(median_scores, labels = device_names,)
plt.show()
```

```python
#As seen above, the pie chart interprets the median scores as sectors of the pie chart

#this time, we'll add the percentage labels to the sectors
plt.pie(median_scores,labels = device_names, autopct='%2.2f%%')
plt.show()
```

```python
#adding the scores and the percentages to each sectors.
plt.pie(median_scores, labels = device_names, autopct=lambda pct: f'{pct:.1f}%\n({int(pct * sum(median_scores) / 100)})')
plt.show()
```

```python
#A swarm plot is a type of categorical scatter plot that displays individual data points along a categorical axis. 
#It is often used to visualize the distribution of data points within different categories or groups. 
#In a swarm plot, the data points are spread out evenly along the categorical axis, avoiding overlap as much as possible.

#next, we use seaborn to plot a swarm plot. we imported the libraries previously so there's no need for a new importation
sb.swarmplot(x='Device Name', y='Median Score', data=sampleset2, hue='Device Name', size=20)
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.xticks(rotation=90, ha='center')
plt.show()
```
