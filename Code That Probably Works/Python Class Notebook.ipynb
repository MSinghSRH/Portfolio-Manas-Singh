```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
years = [1960,1970,1980,1990,2000,2010,2016]
pop = [21.91,28.51,38.67,56.23,66.13,74.57,80.28]
```

Matplotlib

Line plot

with this function we can draw a line with X and Y

always for plotting we should write the **show** function

```python
plt.plot(years, pop)
plt.show
```

```python
plt.figure(figsize=(9,6),dpi=100) #this command is about the window of illustration
plt.plot(years, pop, color='blue') # this command is for plotting
plt.xlabel('Year') #from this line till plt.show it is the setting of the plot (line)
plt.ylabel('Population')
plt.title('Country Population')
plt.yticks([21.91,28.51,38.67,56.23,66.13,74.57,80.28],['21M','28M','38M','56M','66M','74M','80M'])
plt.show()
```

Scatter plot
a good tool for correlation analysis and for identifying clusters or classes

Scatter(x,y)

```python
plt.scatter(years, pop)
plt.show()
```

```python
popsize = np.array(pop)*4
popsize
plt.scatter(years, pop, s=popsize, color='green') #this is a drawing for scatterpoints
plt.xlabel('Year') #from this line till plt.show is the setting of the scatter diagram
plt.ylabel('Population')
plt.title('Country Population')
plt.yticks([21.91,28.51,38.67,56.23,66.13,74.57,80.28],['21M','28M','38M','56M','68M','74M','80M'])
plt.show()
```

```python
popsize = np.array(pop)
popsize
plt.scatter(years, pop, s=popsize, color='green') #this is a drawing for scatterpoints
plt.xlabel('Year') #from this line till plt.show is the setting of the scatter diagram
plt.ylabel('Population')
plt.title('Country Population')
plt.yticks([21.91,28.51,38.67,56.23,66.13,74.57,80.28],['21M','28M','38M','56M','68M','74M','80M'])
plt.show()
```

bar chart

```python
#look at index 4 and 6, which demonstrate overlapping cases.
t1 = [1,3,4,5,6,7,9]
tt1 = [4,7,2,4,7,8,3]

t2 = [2,4,6,8,10]
tt2 = [5,6,2,6,2]

#colors = https://matplotlib.org/api/colors_api.html

plt.bar(t1, tt1, label='Blue bar', color='b')
plt.bar(t2, tt2, label='Green bar', color='g')

plt.xlabel('Bar Number')
plt.ylabel('Bar height')
plt.title('bar chart example')
plt.legend()
plt.show()
```
