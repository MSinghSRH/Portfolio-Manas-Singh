```python
from sklearn import datasets
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

The purpose of this notebook is to develop a supervised machine learning to predict the type of flowers by 4 figures ( parameters) of a flower

In this part we define a dataset with the name of IRIS, so it is not DataFrame from Pandas , however, if we want to see data we have to use the following comand *iris.data*

```python
iris = datasets.load_iris()
```

```python
iris.data.shape #by this command we see the dimensions of the array of iris.data
```

to introduce the data set of iris, we use iris.DESCR

```python
iris.DESCR
```

In this dataset we have two arrays which are:

1- iris.data which includes 4 features of flowers
2- iris.target which includes the type of iris flowers

```python
iris.data
```

```python
iris.feature_names
```

```python
iris_df = pd.DataFrame(iris.data, columns=iris.feature_names)
iris_df
```

Add one new column with the name target which is equal to iris.target

```python
iris_df['target'] = iris.target
iris_df
```

```python
iris_df.isnull().sum()
```

```python
iris.target_names
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=3, metric='minkowski', p=2)
x = iris.data
y = iris.target
knn.fit(x,y)
```

In the prediction, we give the fitures of the new object to the fitted KNN and then the KNN model idetifies the class of this object.

knn.predict gives us the class of each new object based on the fitted data

```python
x_new = np.array([1,1,2,2])
knn.predict([x_new])
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(iris.data, iris.target, test_size=0.3, stratify=iris.target)
```

```python
x_train.shape
```

```python
knn = KNeighborsClassifier(n_neighbors=3)
knn.fit(x_train, y_train)
y_predict = knn.predict(x_test)
y_predict
```

```python
knn.score(x_test, y_test)
```
