```python
import re
hand = open('mbox-short.txt')
numlist = list()
for line in hand:
  line = line.rstrip()
  stuff = re.findall('^X-DSPAM-Confidence: ([0-9.]+)', line)
  if len(stuff) != 1: continue
  num = float(stuff[0])
  numlist.append(num)
print('Maximum:', max(numlist))
```

```python
#how to calculate gross salary of an employee.

#taking inputs for employee classes and number of hours worked
EmpClass = input('Please enter employee class: (1- Management, 2- HR, 3- Staff, 4- Other): ')
Hours = input('Please enter your hours: ')

#input takes string inputs, we need to convert these inputs to integers
EmpClass = int(EmpClass)  
Hours = float(Hours)  

#Conditional statements for calculating salaries
if EmpClass == 1:
  output = Hours*15
  print('Class One: Management')
  print('Your hourly salary is 15 euros')
  print('Your gross salary in Euros is: ', output)
elif EmpClass == 2:
  output = Hours*12
  print('Class Two: Human Resources')
  print('Your hourly salary is 12 euros')
  print('Your gross salary in Euros is: ', output)
elif EmpClass == 3:
  output = Hours*10
  print('Class Three: Staff')
  print('Your hourly salary is 10 euros')
  print('Your gross salary in Euros is: ', output)
elif EmpClass == 4:
  output = Hours*9
  print('Class Four: Other(Including contractors)')
  print('Your hourly salary is 9 euros')
  print('Your gross salary in Euros is: ', output)
else:
  print('Invalid parameters, please contact salary services')
```

```python
# Taking inputs for employee classes and number of hours worked
EmpClass = input('Please enter employee class: (1- Management, 2- HR, 3- Staff, 4- Other): ')
Hours = input('Please enter your hours: ')

# Input takes user input in the form of strings, we need to convert these inputs to integers
EmpClass = int(EmpClass)
Hours = float(Hours)

# Conditional statements for calculating salaries
if EmpClass == 1:
    output = Hours * 15
    if Hours <= 20:
        print('You are a part-time employee')
    else:
        print('You are a full-time employee')
    print('Class One: Management')
    print('Your hourly salary is 15 euros')
    print('Your gross salary is is:', output, 'Euros')
elif EmpClass == 2:
    output = Hours * 12
    if Hours <= 20:
        print('You are a part-time employee')
    else:
        print('You are a full-time employee')
    print('Class Two: Human Resources')
    print('Your hourly salary is 12 euros')
    print('Your gross salary is is:', output, 'Euros')
elif EmpClass == 3:
    output = Hours * 10
    if Hours <= 20:
        print('You are a part-time employee')
    else:
        print('You are a full-time employee')
    print('Class Three: Staff')
    print('Your hourly salary is 10 euros')
    print('Your gross salary is is:', output, 'Euros')
elif EmpClass == 4:
    output = Hours * 9
    if Hours <= 20:
        print('You are a part-time employee')
    else:
        print('You are a full-time employee')
    print('Class Four: Other (Including contractors)')
    print('Your hourly salary is 9 euros')
    print('Your gross salary is is:', output, 'Euros')
else:
    print('Please contact salary services')
```

```python
smallest = None
print('Before')
for value in [3, 41, 12, 9, 74, 15]:
  if smallest is None:
    smallest = value
  elif value < smallest:
    smallest = value
  print(smallest, value)
print('After', smallest)
```
