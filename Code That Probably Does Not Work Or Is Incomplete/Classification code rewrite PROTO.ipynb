```python
#now, we assign the values in the ' Income' column to the feature_names array for use in SciKitLearn
income = ' Income'
#the income_data needs to be reshaped to be able to be used for fitting, because there is only one feature here
#therefore, the income_data array is in 1D format, but we need it in 2D format to be able to fit
#if we try and fit the model using income_data directly, there will be an error
income_data = df[income].values.reshape(-1, 1)
feature_names = [income]
print(feature_names)
```

```python
print(income_data)
```

```python
#now, we assign the values in the ' Class' column to the target array for use in SciKitLearn
income_class = ' Class'
income_class_data = df[income_class].values
target = income_class_data
print(target)
```

```python
#the next step is to map target_names to 0, 1 and 2
#here, 0 corresponds to 'Lower Class', 1 corresponds to 'Middle Class' and 2 corresponds to 'Upper Class'
#we do this so that SciKitLearn can read target names

# Define the target mapping
target_mapping = {0: 'Lower Class', 1: 'Middle Class', 2: 'Upper Class'}

# Map target values to target names
target_names = [target_mapping[value] for value in target]

# Print the target names
print(target_names)
```

```python
#now, we know the target names have been correctly mapped to the numbers 
#we'll use the variables 'income_data' and 'target' to begin fitting
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=3, metric='minkowski',p=2)
knn.fit(income_data, target)
```

```python
#the model has been fitted to accept an income value and make a prediction on the class
#therefore, we now test to check if the ML model works
x_new = np.array([92000])
knn.predict([x_new])
```

```python
#the value 2 corresponds to Upper class
#now, we make a train/test split for our data
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(income_data, target, test_size=0.2,stratify= target)
x_train.shape
```

```python
#now we know the training/testing split (160 entries for train, 40 entries for test, 80/20 split)
#we now fit the model to the training data and ask the model to predict on the test data
knn = KNeighborsClassifier(n_neighbors=3)
knn.fit(x_train, y_train)
y_predict = knn.predict(x_test)
y_predict
```

```python
#finally, we test the model for accuracy.
knn.score(x_test, y_test)
```
